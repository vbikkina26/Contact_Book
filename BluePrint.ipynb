```python
def insert(name,country,number):
    codeQ=f"select phonecode from country where name=='{country}';"
    print(codeQ)
    selCode=db.execute(codeQ)
    db.commit()
    conCode=[i[0] for i in selCode]
#     print(conCode)
    
    insQ=f"insert into example values('{name.capitalize()}',{int(number)},'{country.capitalize()}',{int(conCode[0])});"
    print(insQ)
    db.execute(insQ)
    db.commit()
    
    print(f'Contact {name.capitalize()} Saved successfully !!')

def search(name=None,number=None):
    if name!=None:
        numQ=f"select * from example where name=='{name}';"
        print(numQ)
        selQ=db.execute(numQ)
        db.commit()
        for i in selQ:
            print(f"Name:{i[0]}\nPhoneNumber:+{str(i[3])} {str(i[1])}")

insert(name,country,number)
# search('Nannagaru')
```
